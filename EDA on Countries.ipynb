```python
data['Coastline (coast/area ratio)']=data['Coastline (coast/area ratio)'].str.replace(',','.')
data['Pop. Density (per sq. mi.)']=data['Pop. Density (per sq. mi.)'].str.replace(',','.')
data['Net migration']=data['Net migration'].str.replace(',','.')
data['Infant mortality (per 1000 births)']=data['Infant mortality (per 1000 births)'].str.replace(',','.')
data['Literacy (%)']=data['Literacy (%)'].str.replace(',','.')
data['Phones (per 1000)']=data['Phones (per 1000)'].str.replace(',','.')
data['Arable (%)']=data['Arable (%)'].str.replace(',','.')
data['Crops (%)']=data['Crops (%)'].str.replace(',','.')
data['Other (%)']=data['Other (%)'].str.replace(',','.')
data['Birthrate']=data['Birthrate'].str.replace(',','.')
data['Deathrate']=data['Deathrate'].str.replace(',','.')
data['Agriculture']=data['Agriculture'].str.replace(',','.')
data['Industry']=data['Industry'].str.replace(',','.')
data['Service']=data['Service'].str.replace(',','.')
```

```python
data
```

```python
data.fillna({'Climate':'0'},inplace=True)
data.fillna({'Agriculture':'0'},inplace=True)
data.fillna({'Industry':'0'},inplace=True)
data.fillna({'Service':'0'},inplace=True)
data.fillna({'GDP ($ per capita)':'0'},inplace=True)
data.fillna({'Literacy (%)':'0'},inplace=True)
data.fillna({'Net migration':'0'},inplace=True)
```

```python
data['Birthrate']=data['Birthrate'].astype(float)
data['Coastline (coast/area ratio)']=data['Coastline (coast/area ratio)'].astype(float)
data['Net migration']=data['Net migration'].astype(float)
data['Crops (%)']=data['Crops (%)'].astype(float)
data['Deathrate']=data['Deathrate'].astype(float)
data['Agriculture']=data['Agriculture'].astype(float)
data['Industry']=data['Industry'].astype(float)
data['Service']=data['Service'].astype(float)
data['Literacy (%)']=data['Literacy (%)'].astype(float)
data['Pop. Density (per sq. mi.)']=data['Pop. Density (per sq. mi.)'].astype(float)
data['GDP ($ per capita)']=data['GDP ($ per capita)'].astype(float)
```

```python
data
```

# Top 10 Countries on basis of Population

```python
data.groupby('Country')['Population'].sum().sort_values(ascending=False).head(10).reset_index()
```

```python
data.groupby('Country')['Population'].sum().sort_values(ascending=False).head(10).plot(kind='pie')
```

# Top 10 Countries based on Population Density

```python
data.groupby('Country')['Pop. Density (per sq. mi.)'].sum().sort_values(ascending=False).head(10).reset_index()
```

```python
data.groupby('Country')['Pop. Density (per sq. mi.)'].sum().sort_values(ascending=False).head(10).plot(kind='bar')
```
